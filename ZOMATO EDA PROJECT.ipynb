# Groupby operation

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

# Observation: 

1. When the rating is between 4.5 to 4.9----> Excellent
2. When ratings are between 4.0 to 4.4----> Vry Good
3. When the rating is between 3.5 to 3.9---> Good
4. When the rating is between 2.5 to 3.4---> Average
5. When the rating is between 1.8 to 2.4---> Poor
6. And the white color is showing that the rating is 0.0 show it is not rated
    

```python
ratings.head()
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```

# To increase the size of the graph

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```
